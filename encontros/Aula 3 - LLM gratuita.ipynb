## Teste com DeepSeek R1 - Padrão

```python
prompt_logica = "Se um trem viaja a 80 km/h e precisa percorrer 240 km, mas para por 15 minutos a cada 100 km, quanto tempo levará a viagem total? Pense passo a passo."

client = OpenAI(
    base_url="https://router.huggingface.co/v1",
    api_key=HF_TOKEN,
)
```

```python
completion = client.chat.completions.create(
    model="deepseek-ai/DeepSeek-R1",
    messages=[
        {
            "role": "user","content": prompt_logica
        }
    ],
)
saida = completion.choices[0].message
print(saida.content)
```

```python
display(Markdown(saida.content))
```

##Teste Modelo Qwen 2.5 72B Instruct (modelo atualmente grande)

```python
model_id_qwen = "Qwen/Qwen2.5-72B-Instruct"

def generate_code_qwen(instruction):
    messages = [
        {"role": "system", "content": "Você é um engenheiro de software sênior especialista em Python. a saida deve ser em markdown"},
        {"role": "user", "content": instruction}
    ]

    # Nota: Qwen 2.5 segue instruções de forma muito estrita
    try:
        completion = client.chat.completions.create(
            model=model_id_qwen,
            messages=messages,
            max_tokens=4096, # Maior limite para código extenso
            temperature=0.2, # Temperatura baixa é CRÍTICA para código (precisão > criatividade)
            stream=False     # Sem stream para pegar o bloco completo
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"Erro: {e}"

instrucao_codigo = "Escreva uma API REST completa usando FastAPI que tenha um endpoint para calcular a sequência de Fibonacci com memoization."

print(generate_code_qwen(instrucao_codigo))
```

```python
display(Markdown(generate_code_qwen(instrucao_codigo)))
```

```python
from PyPDF2 import PdfReader
import gradio as gr
```

```python
reader = PdfReader("/content/CV_Machado_Assis.pdf")
cv = """ """
for page in reader.pages:
    text = page.extract_text()
    cv += text
print(cv)
```

```python
with open('/content/cv.txt', 'w') as f:
    f.write(cv)
```

```python
def chat(prompt):
  messages = [
      {"role": "user", "content": prompt},
  ]
  response = client.chat.completions.create(
    model="Qwen/Qwen2.5-72B-Instruct",
    messages=messages
  )
  return response.choices[0].message.content
```

```python
summary_prompt = f"""
faça um resumo do seguinte documento de forma profissional
"""
summary_prompt += cv
summary = chat(summary_prompt)
```

```python
print(summary)
```

```python
with open('/content/summary.txt', 'w') as f:
    f.write(summary)
```

```python
with open('/content/summary.txt', 'r') as f:
    summary = f.read()
```

```python
name = 'Machadinha'
system_prompt = f"""
  você se chama {name} e será responsavel por responder por {name}.
  responda sempre em português.
  você é um profissional engajado e sua responsabilidade é responder um chat
"""
system_prompt += f"""/n você possui a seguinte informação resumida do perfil profissional, use para responder perguntas mais simples: {summary}"""
system_prompt += f"""/n caso alguma informação específica por solicitada, use a fonte de cv completa {cv}"""
```

```python
def chat_pessoal(user_message, history):
    messages = [
        {"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": user_message}]
    try:
        completion = client.chat.completions.create(
            model="Qwen/Qwen2.5-72B-Instruct",
            messages=messages,
            temperature=0.2, # Temperatura baixa é CRÍTICA para código (precisão > criatividade)
            stream=False     # Sem stream para pegar o bloco completo
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"Erro: {e}"
```

```python
gr.ChatInterface(chat_pessoal, type="messages").launch(share=False)
```

```python
def chat_pessoal_gpt(user_message, history):
    messages = [
        {"role": "system", "content": system_prompt}] + history + [{"role": "user", "content": user_message}]
    try:
        completion = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=messages,
            temperature=0.2, # Temperatura baixa é CRÍTICA para código (precisão > criatividade)
            stream=False     # Sem stream para pegar o bloco completo
        )
        return completion.choices[0].message.content
    except Exception as e:
        return f"Erro: {e}"
```

```python
gr.ChatInterface(chat_pessoal_gpt, type="messages").launch(share=False)
```
