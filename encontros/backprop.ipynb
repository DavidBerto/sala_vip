```python
from manim import *

class ArquiteturaRede(Scene):
    def construct(self):
        # 1. Setup Layers
        camada_in = VGroup(*[Circle(radius=0.4, color=BLUE) for _ in range(2)]).arrange(DOWN, buff=1.5).shift(LEFT*4)
        camada_hid = VGroup(*[Circle(radius=0.4, color=WHITE) for _ in range(3)]).arrange(DOWN, buff=1.0)
        camada_out = Circle(radius=0.4, color=GOLD).shift(RIGHT * 4)

        # Labels for nodes
        label_in = VGroup(*[MathTex(f"x_{i+1}").move_to(c) for i, c in enumerate(camada_in)])
        label_hid = VGroup(*[MathTex(f"h_{i+1}").move_to(c) for i, c in enumerate(camada_hid)])
        label_out = MathTex("y").move_to(camada_out)

        # 2. Connections (Weights)
        conexoes_in_hid = VGroup()
        labels_w_in_hid = VGroup()
        for i, c1 in enumerate(camada_in):
            for j, c2 in enumerate(camada_hid):
                line = Line(c1.get_right(), c2.get_left(), stroke_width=2, color=GREY)
                conexoes_in_hid.add(line)
                # Label representative weights (e.g., from x1) to avoid clutter
                if i == 0:
                    lbl = MathTex(f"w^{{(1)}}_{{{j+1}{i+1}}}").scale(0.6)
                    # Adjust position slightly along the line
                    lbl.move_to(line.point_from_proportion(0.25) + UP*0.2)
                    labels_w_in_hid.add(lbl)

        conexoes_hid_out = VGroup()
        labels_w_hid_out = VGroup()
        for j, c1 in enumerate(camada_hid):
            line = Line(c1.get_right(), camada_out.get_left(), stroke_width=2, color=GREY)
            conexoes_hid_out.add(line)
            lbl = MathTex(f"w^{{(2)}}_{{1{j+1}}}").scale(0.6)
            lbl.move_to(line.point_from_proportion(0.25) + UP*0.2)
            labels_w_hid_out.add(lbl)

        # 3. Formulas
        formula_hid = MathTex(r"h_j = \sum x_i w^{(1)}_{ji}").to_edge(UP).scale(0.8)
        formula_out = MathTex(r"y = \sum h_j w^{(2)}_{1j}").to_edge(DOWN).scale(0.8)

        # 4. Animation Sequence
        self.play(Create(camada_in), Write(label_in))
        self.play(Create(camada_hid), Write(label_hid))
        self.play(Create(camada_out), Write(label_out))

        self.play(Create(conexoes_in_hid))
        self.play(Write(labels_w_in_hid))

        self.play(Write(formula_hid))
        self.wait(1)

        self.play(Create(conexoes_hid_out))
        self.play(Write(labels_w_hid_out))

        self.play(Write(formula_out))
        self.wait(2)
```

```python
from manim import *

class FuncoesAtivacao(Scene):
    def construct(self):
        # Título
        titulo = Text("Funções de Ativação").to_edge(UP)
        self.play(Write(titulo))

        # Sistema de Eixos
        axes = Axes(x_range=[-6, 6, 1], y_range=[-1.5, 2.5, 1], x_length=8, y_length=5)
        labels = axes.get_axis_labels(x_label="z", y_label="a")
        self.play(Create(axes), Write(labels))

        # 1. Sigmoide
        sigmoid_func = lambda x: 1 / (1 + np.exp(-x))
        grafico_sigmoid = axes.plot(sigmoid_func, color=BLUE)
        label_sigmoid = MathTex(r"\sigma(z) = \frac{1}{1+e^{-z}}", color=BLUE).to_corner(UL).shift(DOWN)

        self.play(Create(grafico_sigmoid), Write(label_sigmoid))
        self.wait(2)

        # 2. ReLU (Transition)
        relu_func = lambda x: max(0, x)
        grafico_relu = axes.plot(relu_func, color=RED)
        label_relu = MathTex(r"ReLU(z) = \max(0, z)", color=RED).to_corner(UL).shift(DOWN)

        self.play(
            ReplacementTransform(grafico_sigmoid, grafico_relu),
            ReplacementTransform(label_sigmoid, label_relu)
        )
        self.wait(2)

        # Destaque de Não-linearidade
        # texto_final = Text("Introduz Não-Linearidade", font_size=24, color=YELLOW).next_to(axes, DOWN)
        # self.play(Write(texto_final))
        # self.wait()
```

```python
from manim import *

class CalculoNeuronio(Scene):
    def construct(self):
        # Configuração Visual
        # Entradas (x1, x2, x3)
        inputs = VGroup(*[Circle(radius=0.35, color=BLUE) for _ in range(3)]).arrange(DOWN, buff=1.0).shift(LEFT * 4)
        # Bias (visualizado como um nó extra)
        bias = Circle(radius=0.35, color=GRAY).next_to(inputs, UP, buff=1.0)

        neuronio = Circle(radius=0.8, color=WHITE).shift(RIGHT * 1)

        # Labels Entradas e Bias
        x_labels = VGroup(*[MathTex(f"x_{i+1}").move_to(inp) for i, inp in enumerate(inputs)])
        b_label = MathTex("1").move_to(bias) # Entrada do bias é constante 1

        # Linhas de conexão (Inputs -> Neuronio)
        conexoes = VGroup()
        ws = VGroup()
        for i, inp in enumerate(inputs):
            l = Line(inp.get_right(), neuronio.get_left(), color=GREY)
            conexoes.add(l)
            ws.add(MathTex(f"w_{i+1}").scale(0.7).move_to(l.point_from_proportion(0.3)).shift(UP*0.3))

        # Conexão Bias
        l_bias = Line(bias.get_right(), neuronio.get_left(), color=GREY)
        conexoes.add(l_bias)
        w_bias = MathTex("b").scale(0.7).move_to(l_bias.point_from_proportion(0.3)).shift(UP*0.3)
        ws.add(w_bias)

        # 1. Mostrar Estrutura Completa
        self.play(Create(inputs), Write(x_labels), Create(bias), Write(b_label))
        self.play(Create(neuronio))
        self.play(Create(conexoes), Write(ws))
        self.wait()

        # 2. Soma Ponderada
        soma_texto = MathTex(r"z = \sum x_i w_i + b").to_edge(UP)
        self.play(Write(soma_texto))

        # Animação dos sinais fluindo (Inputs + Bias)
        sinais = VGroup(*[Dot(color=YELLOW).move_to(inp) for inp in inputs])
        sinal_bias = Dot(color=YELLOW).move_to(bias)
        sinais.add(sinal_bias)

        self.play(FadeIn(sinais))
        self.play(sinais.animate.move_to(neuronio.get_center()), run_time=1.5)
        self.remove(sinais)

        label_z = MathTex("z").move_to(neuronio)
        self.play(Write(label_z))
        self.wait()

        # 3. Função de Ativação
        ativacao_texto = MathTex(r"a = \sigma(z)").next_to(soma_texto, DOWN)
        self.play(Write(ativacao_texto))

        # Transformação visual (z vira a)
        label_a = MathTex("a").move_to(neuronio)
        self.play(
            Transform(label_z, label_a),
            neuronio.animate.set_fill(BLUE, opacity=0.3)
        )
        self.wait()

        # 4. Saída
        output_arrow = Arrow(neuronio.get_right(), neuronio.get_right() + RIGHT*2, buff=0)
        self.play(GrowArrow(output_arrow))
        y_label = MathTex("y").next_to(output_arrow, RIGHT)
        self.play(Write(y_label))
        self.wait()
```

```python
from manim import *

class ForwardPass(Scene):
    def construct(self):
        # Reaproveitando a estrutura anterior (simplificada)
        n1 = Circle().shift(LEFT * 2)
        n2 = Circle().shift(RIGHT * 2)
        link = Line(n1.get_right(), n2.get_left())
        self.add(n1, n2, link)

        # Animando o sinal (uma pequena esfera de luz)
        sinal = Dot(color=YELLOW).move_to(n1.get_center())

        self.play(sinal.animate.move_to(n2.get_center()), run_time=1.5, rate_func=rate_functions.ease_out_sine)
        self.play(n2.animate.set_fill(YELLOW, opacity=0.5))
        self.wait()
```

```python
from manim import *

class BackwardError(Scene):
    def construct(self):
        # Elementos de texto para a regra da cadeia
        equacao = MathTex(r"\frac{\partial J}{\partial w} = \frac{\partial J}{\partial a} \cdot \frac{\partial a}{\partial z} \cdot \frac{\partial z}{\partial w}")
        equacao.to_edge(UP)

        # Representação visual do erro voltando
        seta_erro = Arrow(RIGHT * 4, LEFT * 4, color=RED)
        label_erro = Text("Erro retropropagado", color=RED).next_to(seta_erro, DOWN)

        self.play(Write(equacao))
        self.play(GrowArrow(seta_erro), FadeIn(label_erro))

        # Destacando o ajuste de um peso específico
        peso = Line(LEFT, RIGHT, stroke_width=4).set_color(WHITE)
        self.add(peso)
        self.play(peso.animate.set_color(RED).set_stroke_width(8)) # O peso "reage" ao erro
        self.play(peso.animate.set_color(GREEN).set_stroke_width(5)) # Peso atualizado e estabilizado
        self.wait()
```
